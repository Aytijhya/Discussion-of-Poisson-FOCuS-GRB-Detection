```python
import numpy as np
import matplotlib.pyplot as plt

def e_detector_sr(data, lmbda):
    N = len(data)
    E = np.zeros(N)
    sum_data = np.cumsum(data)  # Pre-calculate cumulative sum
    l = max(lmbda, (data[0] + lmbda) / 2)
    E[1]= data[0] * np.log(l / lmbda) - l + lmbda
    for i in range(1,N):
        print(i)
        E[i]=E[i-1]
        for j in range(i):
            sum_j_i = sum_data[i-1] - (sum_data[j-1] if j > 0 else 0)
            length = i - j + 1
            l = max(lmbda, (sum_j_i + lmbda) / length)
            E[i] += data[i] * np.log(l / lmbda) - l + lmbda
    return E
```

```python
data = np.random.poisson(300, 200)

plt.plot(e_detector_sr(data, 300))
plt.show()
```

```python
def e_detector_sr(data, lmbda):
    N = len(data)
    E = np.zeros((N, N))
    for j in range(N):
        for i in range(j, N):
            l = max(lmbda, (np.sum(data[j:i])+lmbda)/(i-j+1))
            if i == 0 and j == 0:
                E[j, i] = data[i] * np.log(l / lmbda) - l + lmbda
            else:
                E[j, i] = E[j, i - 1] + data[i] * np.log(l / lmbda) - l + lmbda
    return np.sum(E, axis=0)

data = np.random.poisson(300, 200)

plt.plot(e_detector_sr(data, 300))
plt.show()
```
